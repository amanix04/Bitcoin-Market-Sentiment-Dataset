```python
# Filter for 'BTC'
btc_data = historical_data[historical_data['Coin'] == 'BTC'].copy()
```

```python
# Process fear_greed_index
fear_greed_index['date'] = pd.to_datetime(fear_greed_index['date'])

```

```python
# Merge btc_data with fear_greed_index
btc_merged_data = pd.merge(btc_data, fear_greed_index, on='date', how='left')
```

```python
# Inspect the merged data
print("BTC Merged Data Info:")
btc_merged_data.info()
print("\nBTC Merged Data Head:")
print(btc_merged_data.head())
```

```python
# Save btc_merged_data to csv
btc_merged_data.to_csv('btc_merged_data.csv', index=False)
```
